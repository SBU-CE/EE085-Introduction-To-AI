```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lr.csv')
df.head()
```

```python
plt.scatter(df['x'], df['y'])
```

```python
def predict(X, w, b):
  Y = w*X + b
  return Y

def cost(X, w, b, t):
  Y = [predict(x, w, b) for x in X]
  N = len(X)
  cost = np.sum((Y - t) ** 2) / (2*N)
  return cost
```

```python
def gradient_descent(X, t):
  w = 0
  b = 0
  alpha = 0.002
  N = len(X)

  for i in range(100000):
    Y = [predict(x, w, b) for x in X]
    w = w - (alpha/N) * np.sum(np.multiply(Y-t, X))
    b = b - (alpha/N) * np.sum(Y-t)

    if i%100 ==0:
      print(cost(X, w, b, t))
  return w, b
```

```python
best_w, best_b = gradient_descent(df['x'], df['y'])
```

```python
y_predict = predict(df['x'], best_w, best_b)
plt.scatter(df['x'], df['y'])
plt.plot(df['x'], y_predict, color='r')
```
