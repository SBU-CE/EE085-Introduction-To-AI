## Read Dataset

```python
data = pd.read_csv('/content/Bank Customer Churn Prediction.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

## Separate Features and Targets

```python
y = data.iloc[:,-1].values
```

```python
y
```

## PreProcessing

```python
data['gender'].unique()
```

### Drop not important Columns

```python
data.drop(['customer_id', 'churn'], axis=1, inplace=True)
```

```python
data.head()
```

### Convert Categorical into Numeric

#### One-Hot Encoding

```python
data['country'].unique()
```

```python
pd.get_dummies(data, columns=['country'], prefix='country')
```

```python
data = pd.get_dummies(data, columns=['country'], prefix='country')
```

#### Label Encoding

```python
data['gender'] = data['gender'].apply(lambda x: 1 if x == 'Male' else 0)
```

```python
data.head()
```

```python
x = data.values
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x
```

```python
print(x.shape)
print(y.shape)
```

### Split Dataset

$0.6$ train, $0.2$ validation, and $0.2$ test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```

### Model Selection

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_train = lr.predict(X_train)
y_pred_val = lr.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score

print(f"Train accuracy: {accuracy_score(y_train, y_pred_train)}")
print(f"Val accuracy: {accuracy_score(y_val, y_pred_val)}")
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier().fit(X_train, y_train)
```

```python
y_pred_train = rfc.predict(X_train)
y_pred_val = rfc.predict(X_val)
```

```python
print(f"Train accuracy: {accuracy_score(y_train, y_pred_train)}")
print(f"Val accuracy: {accuracy_score(y_val, y_pred_val)}")
```
